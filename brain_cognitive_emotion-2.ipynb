```python
import scipy.io
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 3: Preprocess Data
print("Step 3: Preprocessing EEG and EOG...")
eeg_file_path = '/content/eeg_features.mat'
eog_file_path = '/content/eog_features.mat'  # TODO: Replace with actual path
try:
    eeg_data = scipy.io.loadmat(eeg_file_path)
    print("EEG file loaded:", eeg_file_path)
except FileNotFoundError:
    print(f"Error: {eeg_file_path} not found.")
    exit()

try:
    eog_data = scipy.io.loadmat(eog_file_path)
    print("EOG file loaded:", eog_file_path)
except FileNotFoundError:
    print("Warning: EOG file not found. Using EEG only.")
    eog_data = {}

eeg_vars = ['psd_movingAve', 'psd_LDS', 'de_movingAve', 'de_LDS']
eog_vars = ['features_table_ica']
total_samples = 885

eeg_dfs = []
for var in eeg_vars:
    if var not in eeg_data:
        print(f"Warning: {var} not found.")
        continue
    data = eeg_data[var]
    if data.shape != (17, 885, 5):
        print(f"Warning: {var} shape {data.shape}, expected (17, 885, 5).")
        continue
    data = data.transpose(1, 0, 2).reshape(885, 17 * 5)
    print(f"Reshaped {var} to {data.shape}")
    columns = [f"{var}_ch{c}_band{b}" for c in range(17) for b in range(5)]
    df = pd.DataFrame(data, columns=columns)
    eeg_dfs.append(df)

eog_dfs = []
for var in eog_vars:
    if var not in eog_data:
        print(f"Warning: {var} not found.")
        continue
    data = eog_data[var]
    if data.shape != (885, 36):
        print(f"Warning: {var} shape {data.shape}, expected (885, 36).")
        continue
    print(f"{var} shape: {data.shape}")
    columns = [f"{var}_{i}" for i in range(36)]
    df = pd.DataFrame(data, columns=columns)
    eog_dfs.append(df)

all_dfs = eeg_dfs + eog_dfs
if not all_dfs:
    print("Error: No features found.")
    exit()
df_features = pd.concat(all_dfs, axis=1)
print("Combined features shape:", df_features.shape)

scaler = StandardScaler()
df_features = pd.DataFrame(scaler.fit_transform(df_features), columns=df_features.columns)
df_features.to_csv('preprocessed_features.csv', index=False)
print("Saved 'preprocessed_features.csv'")
```

```python
# Step 4: Clean Data
print("\nStep 4: Cleaning...")
df_features = pd.read_csv('preprocessed_features.csv')
print("Missing values before:", df_features.isna().sum().sum())
df_features = df_features.fillna(df_features.mean(numeric_only=True))
print("Missing values after:", df_features.isna().sum().sum())

z_scores = np.abs((df_features - df_features.mean(numeric_only=True)) /
                  df_features.std(numeric_only=True))
df_clean = df_features[(z_scores < 3).all(axis=1)]
print("Shape before outlier removal:", df_features.shape)
print("Shape after outlier removal:", df_clean.shape)
df_clean.to_csv('cleaned_features.csv', index=False)
print("Saved 'cleaned_features.csv'")
```

```python
# Step 5: Extract Labels
print("\nStep 5: Extracting Labels...")
label_file_path = '/content/labels.mat'  # TODO: Confirm path
try:
    label_data = scipy.io.loadmat(label_file_path)
    print("Label file loaded:", label_file_path)
except FileNotFoundError:
    print(f"Error: {label_file_path} not found.")
    exit()

label_var = 'perclos'
if label_var in label_data:
    labels = label_data[label_var].flatten()
    print(f"Found {label_var}")
    if len(labels) != total_samples:
        print(f"Warning: Label count {len(labels)} != {total_samples}")
        if len(labels) < total_samples:
            labels = np.pad(labels, (0, total_samples - len(labels)), mode='edge')
        else:
            labels = labels[:total_samples]
else:
    print("Error: perclos not found.")
    labels = np.zeros(total_samples)

df_labels = pd.DataFrame({'vigilance': labels})
print("Vigilance shape:", labels.shape)
print("Vigilance range: {:.4f} to {:.4f}".format(labels.min(), labels.max()))

f3_ch, f4_ch = 3, 4  # F3, F4 per 10-20 system (not FP1=1, FP2=2)
alpha_band = 2  # 8-14 Hz
if 'de_movingAve' in eeg_data:
    eeg = eeg_data['de_movingAve']
    if eeg.shape == (17, 885, 5):
        f3_alpha = eeg[f3_ch-1, :, alpha_band]
        f4_alpha = eeg[f4_ch-1, :, alpha_band]
        print("F3 alpha shape:", f3_alpha.shape)
        print("F4 alpha shape:", f4_alpha.shape)
        valence_score = f4_alpha - f3_alpha
        print("Valence score range: {:.4f} to {:.4f}".format(valence_score.min(), valence_score.max()))
        print("Valence score std: {:.4f}".format(valence_score.std()))
        if abs(valence_score).max() < 0.05:
            print("Warning: Small valence scores. Normalizing...")
            f3_alpha = (f3_alpha - f3_alpha.mean()) / f3_alpha.std()
            f4_alpha = (f4_alpha - f4_alpha.mean()) / f4_alpha.std()
            valence_score = f4_alpha - f3_alpha
            print("Normalized valence range: {:.4f} to {:.4f}".format(valence_score.min(), valence_score.max()))
    else:
        print("Invalid EEG shape:", eeg.shape)
        f3_alpha = np.zeros(total_samples)
        f4_alpha = np.zeros(total_samples)
else:
    print("de_movingAve not found.")
    f3_alpha = np.zeros(total_samples)
    f4_alpha = np.zeros(total_samples)

def get_valence(f3_alpha, f4_alpha, threshold=0.05):
    valence_score = f4_alpha - f3_alpha
    valence = np.where(valence_score > threshold, 'Positive',
                      np.where(valence_score < -threshold, 'Negative', 'Neutral'))
    return valence

def get_arousal(perclos):
    return np.where(perclos < 0.3, 'High',
                   np.where(perclos < 0.6, 'Medium', 'Low'))

def map_to_emotion(valence, arousal):
    if valence == 'Positive' and arousal == 'High':
        return 'Happy'
    elif valence == 'Positive' and arousal == 'Medium':
        return 'Relaxed'
    elif valence == 'Positive' and arousal == 'Low':
        return 'Calm'
    elif valence == 'Negative' and arousal == 'High':
        return 'Angry'
    elif valence == 'Negative' and arousal == 'Medium':
        return 'Frustrated'
    elif valence == 'Negative' and arousal == 'Low':
        return 'Sad'
    else:
        return 'Neutral'

threshold = 0.05  # Try 0.1 if Negative-heavy
df_labels['valence'] = get_valence(f3_alpha, f4_alpha, threshold)
df_labels['arousal'] = get_arousal(labels)
df_labels['emotion'] = [map_to_emotion(v, a) for v, a in zip(df_labels['valence'], df_labels['arousal'])]
df_labels['binary_vigilance'] = (labels > 0.5).astype(int)

print("\nBinary vigilance distribution:")
print(df_labels['binary_vigilance'].value_counts())
print("\nValence distribution:")
print(df_labels['valence'].value_counts())
print("\nArousal distribution:")
print(df_labels['arousal'].value_counts())
print("\nEmotion distribution:")
print(df_labels['emotion'].value_counts(normalize=True))
if df_labels['emotion'].value_counts().min() < 50:
    print("Warning: Imbalanced emotions (e.g., Calm <50). Using SMOTE in training.")

df_labels.to_csv('extracted_labels.csv', index=False)
print("Labels saved as 'extracted_labels.csv'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_val_predict
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, Dropout, LSTM, Attention, Flatten, Dense
from tensorflow.keras.callbacks import EarlyStopping

# Step 1: Load and Prepare Data
print("\nStep 1: Loading Data...")
df = pd.read_csv('emotion_classified_seed_vig_dataset.csv')
feature_cols = [col for col in df.columns if col.startswith(('psd', 'de', 'features_table_'))]
X = df[feature_cols].values
y = df['emotion'].values

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)
print("Encoded emotions:", le.classes_)

# Standardize features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Balance the dataset using SMOTE
smote = SMOTE(random_state=42)
X, y_encoded = smote.fit_resample(X, y_encoded)
print("After SMOTE, shape:", X.shape)

# Reshape for CNN-LSTM
X_cnn_lstm = X.reshape(X.shape[0], 1, X.shape[1])

# Step 2: Train Classical ML Models
print("\nStep 2: Training Models...")
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42, n_jobs=-1),
    'SVM': SVC(kernel='rbf', class_weight='balanced', probability=True, random_state=42),
    'XGBoost': XGBClassifier(random_state=42, n_jobs=-1, use_label_encoder=False, eval_metric='mlogloss'),
    'MLP': MLPClassifier(hidden_layer_sizes=(128, 64), max_iter=500, random_state=42)
}

results = {}
for name, model in models.items():
    scores = cross_val_score(model, X, y_encoded, cv=cv, scoring='accuracy', n_jobs=-1)
    results[name] = scores
    print(f"{name} CV accuracy: {scores.mean():.4f}  {scores.std():.4f}")

# Step 3: CNN-LSTM-Attention Model
print("\nTraining CNN-LSTM-Attention with OOF predictions for stacking...")
def create_cnn_lstm_attention(input_shape, num_classes):
    inputs = Input(shape=input_shape)
    conv = Conv1D(64, kernel_size=3, activation='relu', padding='same')(inputs)
    conv = Dropout(0.3)(conv)
    lstm = LSTM(64, return_sequences=True)(conv)
    attention = Attention()([lstm, lstm])
    flat = Flatten()(attention)
    dense = Dense(128, activation='relu')(flat)
    dense = Dropout(0.3)(dense)
    outputs = Dense(num_classes, activation='softmax')(dense)
    return Model(inputs, outputs)

cnn_preds = np.zeros((X.shape[0], len(le.classes_)))
rf_preds = np.zeros((X.shape[0], len(le.classes_)))

for train_idx, val_idx in cv.split(X, y_encoded):
    # CNN-LSTM
    X_train, X_val = X_cnn_lstm[train_idx], X_cnn_lstm[val_idx]
    y_train = y_encoded[train_idx]

    model = create_cnn_lstm_attention((1, X.shape[1]), len(le.classes_))
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=0, validation_split=0.1, callbacks=[EarlyStopping(patience=3)])
    cnn_preds[val_idx] = model.predict(X_val, verbose=0)

    # RF
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
    rf_model.fit(X[train_idx], y_encoded[train_idx])
    rf_preds[val_idx] = rf_model.predict_proba(X[val_idx])

# Step 4: Meta Model (Stacked Ensemble)
print("\nTraining Stacked Ensemble with Logistic Regression...")
meta_X = np.hstack([cnn_preds, rf_preds])
meta_model = LogisticRegression(max_iter=500)
meta_model.fit(meta_X, y_encoded)
final_preds = meta_model.predict(meta_X)
stacked_accuracy = accuracy_score(y_encoded, final_preds)
print(f"Stacked Ensemble accuracy (OOF): {stacked_accuracy:.4f}")

# Step 5: Evaluation
print("\nStep 5: Evaluating Base Models...")
for name, model in models.items():
    y_pred = cross_val_predict(model, X, y_encoded, cv=cv, n_jobs=-1)
    print(f"\n{name} Classification Report:")
    print(classification_report(y_encoded, y_pred, target_names=le.classes_))
    cm = confusion_matrix(y_encoded, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
    plt.title(f'{name} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Evaluate Stacked Ensemble
print("\nStacked Ensemble Classification Report:")
print(classification_report(y_encoded, final_preds, target_names=le.classes_))
cm = confusion_matrix(y_encoded, final_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Stacked Ensemble Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import numpy as np
import pandas as pd
import keras

# Define constants
NUM_CLASSES = 7  # Seven emotion classes
ADAPTER_DIM = 16  # Bottleneck dimension for adapter layers
N_SPLITS = 5  # For 5-fold cross-validation
EMOTION_MAPPING = {
    "Angry": 0,
    "Calm": 1,
    "Frustrated": 2,
    "Happy": 3,
    "Neutral": 4,
    "Relaxed": 5,
    "Sad": 6
}

# Custom Attention Layer
class AttentionLayer(layers.Layer):
    def __init__(self, name="attention"):
        super(AttentionLayer, self).__init__(name=name)

    def call(self, inputs):
        lstm_output, attention_weights = inputs
        weighted_output = lstm_output * attention_weights
        context_vector = keras.ops.sum(weighted_output, axis=1)
        return context_vector

# Adapter Layer (Bottleneck MLP)
class AdapterLayer(layers.Layer):
    def __init__(self, input_dim, bottleneck_dim, name="adapter"):
        super(AdapterLayer, self).__init__(name=name)
        self.down_project = layers.Dense(bottleneck_dim, activation="relu")
        self.up_project = layers.Dense(input_dim)
        self.batch_norm = layers.BatchNormalization()

    def call(self, inputs, training=False):
        x = self.down_project(inputs)
        x = self.batch_norm(x, training=training)
        x = self.up_project(x)
        return x + inputs

# Custom Weighted Adapter Layer
class WeightedAdapterLayer(layers.Layer):
    def __init__(self, time_steps, name="weighted_adapter"):
        super(WeightedAdapterLayer, self).__init__(name=name)
        self.time_steps = time_steps

    def call(self, inputs):
        adapter_out, weight = inputs
        weight_expanded = keras.ops.expand_dims(weight, axis=1)
        weight_expanded = keras.ops.repeat(weight_expanded, self.time_steps, axis=1)
        return adapter_out * weight_expanded

# Gating Mechanism
class GatingMechanism(layers.Layer):
    def __init__(self, num_classes, num_adapters, name="gating"):
        super(GatingMechanism, self).__init__(name=name)
        self.num_classes = num_classes
        self.num_adapters = num_adapters
        self.emotion_predictor = layers.Dense(num_classes, activation="softmax", name="emotion_predictor")
        self.adapter_selector = layers.Dense(num_adapters, activation="softmax", name="adapter_selector")

    def call(self, inputs, training=False):
        emotion_dist = self.emotion_predictor(inputs)
        adapter_weights = self.adapter_selector(emotion_dist)
        return adapter_weights

# CNN-LSTM-Attention Model with Adapters
def build_cnn_lstm_attention_with_adapters(input_shape, num_classes=NUM_CLASSES):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv1D(filters=64, kernel_size=3, padding="same", activation="relu")(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    reduced_time_steps = input_shape[0] // 2
    cnn_adapters = [AdapterLayer(input_dim=64, bottleneck_dim=ADAPTER_DIM, name=f"cnn_adapter_{i}") for i in range(num_classes)]
    cnn_adapter_outputs = [adapter(x) for adapter in cnn_adapters]
    cnn_gating_input = layers.GlobalAveragePooling1D()(x)
    cnn_gating = GatingMechanism(num_classes=num_classes, num_adapters=num_classes, name="cnn_gating")
    cnn_adapter_weights = cnn_gating(cnn_gating_input)
    cnn_output = 0
    for i, adapter_out in enumerate(cnn_adapter_outputs):
        weighted_out = WeightedAdapterLayer(time_steps=reduced_time_steps, name=f"cnn_weighted_adapter_{i}")(
            [adapter_out, cnn_adapter_weights[:, i:i+1]]
        )
        cnn_output += weighted_out
    x = layers.LSTM(128, return_sequences=True)(cnn_output)
    x = layers.BatchNormalization()(x)
    lstm_adapters = [AdapterLayer(input_dim=128, bottleneck_dim=ADAPTER_DIM, name=f"lstm_adapter_{i}") for i in range(num_classes)]
    lstm_adapter_outputs = [adapter(x) for adapter in lstm_adapters]
    lstm_gating_input = layers.GlobalAveragePooling1D()(x)
    lstm_gating = GatingMechanism(num_classes=num_classes, num_adapters=num_classes, name="lstm_gating")
    lstm_adapter_weights = lstm_gating(lstm_gating_input)
    lstm_output = 0
    for i, adapter_out in enumerate(lstm_adapter_outputs):
        weighted_out = WeightedAdapterLayer(time_steps=reduced_time_steps, name=f"lstm_weighted_adapter_{i}")(
            [adapter_out, lstm_adapter_weights[:, i:i+1]]
        )
        lstm_output += weighted_out
    attention_scores = layers.Dense(1, activation="tanh")(lstm_output)
    attention_weights = layers.Softmax(axis=1)(attention_scores)
    context_vector = AttentionLayer()([lstm_output, attention_weights])
    x = layers.Dense(128, activation="relu")(context_vector)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = Model(inputs, outputs)
    return model

# Function to find a suitable TIME_STEPS
def find_time_steps(num_features, preferred_divisors=[85, 94, 47, 25, 5, 4]):
    for divisor in preferred_divisors:
        if num_features % divisor == 0:
            return divisor
    for i in range(1, num_features + 1):
        if num_features % i == 0:
            return i
    raise ValueError(f"No valid TIME_STEPS found for {num_features} features.")

# Load and preprocess SEED-VIG dataset from CSV
def load_seed_vig_data(file_path="emotion_classified_seed_vig_dataset (2).csv"):
    try:
        data = pd.read_csv(file_path)
    except FileNotFoundError:
        raise FileNotFoundError(f"Dataset file {file_path} not found. Ensure the SEED-VIG CSV is in the working directory.")

    if "emotion" not in data.columns:
        raise ValueError("CSV must contain an 'emotion' column with string labels (Angry, Calm, Frustrated, Happy, Neutral, Relaxed, Sad).")

    if not all(data["emotion"].isin(EMOTION_MAPPING.keys())):
        invalid_labels = set(data["emotion"]) - set(EMOTION_MAPPING.keys())
        raise ValueError(f"Invalid emotion labels found: {invalid_labels}. Expected: {list(EMOTION_MAPPING.keys())}.")

    y = data["emotion"].map(EMOTION_MAPPING).values
    print(f"Emotion distribution:\n{data['emotion'].value_counts()}")

    # Extract feature columns
    feature_columns = [col for col in data.columns if col not in ["vigilance", "valence", "arousal", "emotion", "binary_vigilance"]]
    num_features = len(feature_columns)
    print(f"Detected {num_features} feature columns: {feature_columns[:10]}...")
    print(f"Total columns in CSV: {len(data.columns)}")
    X = data[feature_columns].values
    print(f"Feature matrix shape: {X.shape}")

    # Check for expected EEG and ICA features
    expected_eeg = [f"{feat}_ch{ ch}_band{band}" for feat in ["psd_movingAve", "psd_LDS", "de_movingAve", "de_LDS"]
                    for ch in range(17) for band in range(5)]
    expected_ica = [f"features_table_ica_{i}" for i in range(36)]
    expected_features = expected_eeg + expected_ica  # 340 + 36 = 425
    missing_features = [col for col in expected_features if col not in feature_columns]
    print(f"Missing {len(missing_features)} expected features: {missing_features[:10]}..." if missing_features else "All expected features present.")
    if num_features != 425:
        print(f"WARNING: Expected 425 features, but found {num_features}. Verify CSV content.")

    # Dynamically select TIME_STEPS
    time_steps = find_time_steps(num_features)
    print(f"Selected TIME_STEPS={time_steps} for {num_features} features.")

    from sklearn.impute import SimpleImputer
    imputer = SimpleImputer(strategy="mean")
    X = imputer.fit_transform(X)
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    smote = SMOTE(random_state=42)
    X_balanced, y_balanced = smote.fit_resample(X, y)
    features_per_step = num_features // time_steps
    X_reshaped = X_balanced.reshape(-1, time_steps, features_per_step)
    y_one_hot = tf.keras.utils.to_categorical(y_balanced, num_classes=NUM_CLASSES)
    print(f"Loaded data: {X_reshaped.shape} samples, {y_one_hot.shape} labels")
    print(f"Input shape: ({time_steps}, {features_per_step})")
    print(f"Balanced class distribution: {np.sum(y_one_hot, axis=0)}")
    return X_reshaped, y_one_hot, features_per_step, time_steps

# Evaluate model with cross-validation
def evaluate_model(X, y, input_shape, num_classes=NUM_CLASSES, n_splits=N_SPLITS):
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    accuracies = []
    precisions, recalls, f1s = [], [], []  # Fixed initialization
    y_classes = y.argmax(axis=1)
    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y_classes)):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        model = build_cnn_lstm_attention_with_adapters(input_shape)
        model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
        model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2, verbose=1)
        y_pred = model.predict(X_test)
        y_pred_classes = y_pred.argmax(axis=1)
        y_test_classes = y_test.argmax(axis=1)
        acc = accuracy_score(y_test_classes, y_pred_classes)
        prec, rec, f1, _ = precision_recall_fscore_support(y_test_classes, y_pred_classes, average="weighted")
        accuracies.append(acc)
        precisions.append(prec)
        recalls.append(rec)
        f1s.append(f1)
        print(f"Fold {fold+1}: Accuracy = {acc:.4f}, Precision = {prec:.4f}, Recall = {rec:.4f}, F1 = {f1:.4f}")
    mean_acc = np.mean(accuracies)
    std_acc = np.std(accuracies)
    print(f"\nMean Accuracy: {mean_acc:.4f}  {std_acc:.4f}")
    print(f"Mean Precision: {np.mean(precisions):.4f}")
    print(f"Mean Recall: {np.mean(recalls):.4f}")
    print(f"Mean F1-Score: {np.mean(f1s):.4f}")
    return mean_acc, precisions, recalls, f1s

# Main execution
if __name__ == "__main__":
    # Load SEED-VIG data
    try:
        X, y, features_per_step, time_steps = load_seed_vig_data("emotion_classified_seed_vig_dataset (2).csv")
        input_shape = (time_steps, features_per_step)
        print(f"Final input shape: {input_shape}")
    except Exception as e:
        print(f"Error loading data: {e}")
        raise ValueError("Failed to load CSV. Check file path, feature count, and column structure.")

    # Evaluate adapter-enhanced model
    print("Evaluating Adapter-Enhanced CNN-LSTM-Attention Model")
    mean_acc, precisions, recalls, f1s = evaluate_model(X, y, input_shape)

    # Compare with original study baselines
    baseline_models = {
        "SVM": 92.32,
        "Random Forest": 95.41,
        "XGBoost": 96.41,
        "MLP": 97.14,
        "Stacked Ensemble": 95.56
    }
    print("\nBaseline Accuracies from Original Study:")
    for model, acc in baseline_models.items():
        print(f"{model}: {acc:.2f}%")
    print(f"Adapter-Enhanced Model: {mean_acc*100:.2f}%")
```

```python

import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import numpy as np
import pandas as pd
import keras

# Define constants
NUM_CLASSES = 7  # Seven emotion classes
ADAPTER_DIM = 16  # Bottleneck dimension for adapter layers
N_SPLITS = 5  # For 5-fold cross-validation
EMOTION_MAPPING = {
    "Angry": 0,
    "Calm": 1,
    "Frustrated": 2,
    "Happy": 3,
    "Neutral": 4,
    "Relaxed": 5,
    "Sad": 6
}

# Custom Attention Layer
class AttentionLayer(layers.Layer):
    def __init__(self, name="attention"):
        super(AttentionLayer, self).__init__(name=name)

    def call(self, inputs):
        lstm_output, attention_weights = inputs
        weighted_output = lstm_output * attention_weights
        context_vector = keras.ops.sum(weighted_output, axis=1)
        return context_vector

# Adapter Layer (Bottleneck MLP)
class AdapterLayer(layers.Layer):
    def __init__(self, input_dim, bottleneck_dim, name="adapter"):
        super(AdapterLayer, self).__init__(name=name)
        self.down_project = layers.Dense(bottleneck_dim, activation="relu")
        self.up_project = layers.Dense(input_dim)
        self.batch_norm = layers.BatchNormalization()

    def call(self, inputs, training=False):
        x = self.down_project(inputs)
        x = self.batch_norm(x, training=training)
        x = self.up_project(x)
        return x + inputs

# Custom Weighted Adapter Layer
class WeightedAdapterLayer(layers.Layer):
    def __init__(self, time_steps, name="weighted_adapter"):
        super(WeightedAdapterLayer, self).__init__(name=name)
        self.time_steps = time_steps

    def call(self, inputs):
        adapter_out, weight = inputs
        weight_expanded = keras.ops.expand_dims(weight, axis=1)
        weight_expanded = keras.ops.repeat(weight_expanded, self.time_steps, axis=1)
        return adapter_out * weight_expanded

# Gating Mechanism
class GatingMechanism(layers.Layer):
    def __init__(self, num_classes, num_adapters, name="gating"):
        super(GatingMechanism, self).__init__(name=name)
        self.num_classes = num_classes
        self.num_adapters = num_adapters
        self.emotion_predictor = layers.Dense(num_classes, activation="softmax", name="emotion_predictor")
        self.adapter_selector = layers.Dense(num_adapters, activation="softmax", name="adapter_selector")

    def call(self, inputs, training=False):
        emotion_dist = self.emotion_predictor(inputs)
        adapter_weights = self.adapter_selector(emotion_dist)
        return adapter_weights

# CNN-LSTM-Attention Model with Adapters
def build_cnn_lstm_attention_with_adapters(input_shape, num_classes=NUM_CLASSES):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv1D(filters=64, kernel_size=3, padding="same", activation="relu")(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    reduced_time_steps = input_shape[0] // 2
    cnn_adapters = [AdapterLayer(input_dim=64, bottleneck_dim=ADAPTER_DIM, name=f"cnn_adapter_{i}") for i in range(num_classes)]
    cnn_adapter_outputs = [adapter(x) for adapter in cnn_adapters]
    cnn_gating_input = layers.GlobalAveragePooling1D()(x)
    cnn_gating = GatingMechanism(num_classes=num_classes, num_adapters=num_classes, name="cnn_gating")
    cnn_adapter_weights = cnn_gating(cnn_gating_input)
    cnn_output = 0
    for i, adapter_out in enumerate(cnn_adapter_outputs):
        weighted_out = WeightedAdapterLayer(time_steps=reduced_time_steps, name=f"cnn_weighted_adapter_{i}")(
            [adapter_out, cnn_adapter_weights[:, i:i+1]]
        )
        cnn_output += weighted_out
    x = layers.LSTM(128, return_sequences=True)(cnn_output)
    x = layers.BatchNormalization()(x)
    lstm_adapters = [AdapterLayer(input_dim=128, bottleneck_dim=ADAPTER_DIM, name=f"lstm_adapter_{i}") for i in range(num_classes)]
    lstm_adapter_outputs = [adapter(x) for adapter in lstm_adapters]
    lstm_gating_input = layers.GlobalAveragePooling1D()(x)
    lstm_gating = GatingMechanism(num_classes=num_classes, num_adapters=num_classes, name="lstm_gating")
    lstm_adapter_weights = lstm_gating(lstm_gating_input)
    lstm_output = 0
    for i, adapter_out in enumerate(lstm_adapter_outputs):
        weighted_out = WeightedAdapterLayer(time_steps=reduced_time_steps, name=f"lstm_weighted_adapter_{i}")(
            [adapter_out, lstm_adapter_weights[:, i:i+1]]
        )
        lstm_output += weighted_out
    attention_scores = layers.Dense(1, activation="tanh")(lstm_output)
    attention_weights = layers.Softmax(axis=1)(attention_scores)
    context_vector = AttentionLayer()([lstm_output, attention_weights])
    x = layers.Dense(128, activation="relu")(context_vector)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = Model(inputs, outputs)
    return model

# Function to find a suitable TIME_STEPS
def find_time_steps(num_features, preferred_divisors=[94, 47, 4, 85, 25, 5]):
    for divisor in preferred_divisors:
        if num_features % divisor == 0:
            return divisor
    for i in range(1, num_features + 1):
        if num_features % i == 0:
            return i
    raise ValueError(f"No valid TIME_STEPS found for {num_features} features.")

# Load and preprocess SEED-VIG dataset from CSV
def load_seed_vig_data(file_path="emotion_classified_seed_vig_dataset (2).csv"):
    try:
        data = pd.read_csv(file_path)
    except FileNotFoundError:
        raise FileNotFoundError(f"Dataset file {file_path} not found. Ensure the SEED-VIG CSV is in the working directory.")

    if "emotion" not in data.columns:
        raise ValueError("CSV must contain an 'emotion' column with string labels (Angry, Calm, Frustrated, Happy, Neutral, Relaxed, Sad).")

    if not all(data["emotion"].isin(EMOTION_MAPPING.keys())):
        invalid_labels = set(data["emotion"]) - set(EMOTION_MAPPING.keys())
        raise ValueError(f"Invalid emotion labels found: {invalid_labels}. Expected: {list(EMOTION_MAPPING.keys())}.")

    y = data["emotion"].map(EMOTION_MAPPING).values
    print(f"Emotion distribution (raw):\n{data['emotion'].value_counts()}")

    # Extract feature columns
    feature_columns = [col for col in data.columns if col not in ["vigilance", "valence", "arousal", "emotion", "binary_vigilance"]]
    num_features = len(feature_columns)
    print(f"Detected {num_features} feature columns: {feature_columns[:10]}...")
    print(f"Total columns in CSV: {len(data.columns)}")
    X = data[feature_columns].values
    print(f"Feature matrix shape: {X.shape}")

    # Check for expected EEG and ICA features
    expected_eeg = [f"{feat}_ch{ch}_band{band}" for feat in ["psd_movingAve", "psd_LDS", "de_movingAve", "de_LDS"]
                    for ch in range(17) for band in range(5)]
    expected_ica = [f"features_table_ica_{i}" for i in range(36)]
    expected_features = expected_eeg + expected_ica  # 340 + 36 = 425
    missing_features = [col for col in expected_features if col not in feature_columns]
    print(f"Missing {len(missing_features)} expected features: {missing_features[:10]}..." if missing_features else "All expected features present.")
    if num_features != 425:
        print(f"WARNING: Expected 425 features, but found {num_features}. Proceeding with {num_features} features.")

    # Dynamically select TIME_STEPS
    time_steps = find_time_steps(num_features)
    print(f"Selected TIME_STEPS={time_steps} for {num_features} features.")

    from sklearn.impute import SimpleImputer
    imputer = SimpleImputer(strategy="mean")
    X = imputer.fit_transform(X)
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    smote = SMOTE(random_state=42, sampling_strategy="auto")
    X_balanced, y_balanced = smote.fit_resample(X, y)
    features_per_step = num_features // time_steps
    X_reshaped = X_balanced.reshape(-1, time_steps, features_per_step)
    y_one_hot = tf.keras.utils.to_categorical(y_balanced, num_classes=NUM_CLASSES)
    print(f"Loaded data: {X_reshaped.shape} samples, {y_one_hot.shape} labels")
    print(f"Input shape: ({time_steps}, {features_per_step})")
    print(f"Balanced class distribution: {np.sum(y_one_hot, axis=0)}")
    return X_reshaped, y_one_hot, features_per_step, time_steps

# Evaluate model with cross-validation
def evaluate_model(X, y, input_shape, num_classes=NUM_CLASSES, n_splits=N_SPLITS):
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    accuracies = []
    precisions, recalls, f1s = [], [], []
    y_classes = y.argmax(axis=1)
    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y_classes)):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        model = build_cnn_lstm_attention_with_adapters(input_shape)
        model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
        # Callbacks for optimization
        early_stopping = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
        lr_scheduler = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6)
        model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2,
                  callbacks=[early_stopping, lr_scheduler], verbose=1)
        y_pred = model.predict(X_test)
        y_pred_classes = y_pred.argmax(axis=1)
        y_test_classes = y_test.argmax(axis=1)
        acc = accuracy_score(y_test_classes, y_pred_classes)
        prec, rec, f1, _ = precision_recall_fscore_support(y_test_classes, y_pred_classes, average="weighted")
        # Per-class F1-scores
        prec_per_class, rec_per_class, f1_per_class, _ = precision_recall_fscore_support(
            y_test_classes, y_pred_classes, labels=range(num_classes)
        )
        print(f"Fold {fold+1}: Accuracy = {acc:.4f}, Precision = {prec:.4f}, Recall = {rec:.4f}, F1 = {f1:.4f}")
        print("Per-class F1-scores:")
        for i, emotion in enumerate(EMOTION_MAPPING.keys()):
            print(f"  {emotion}: F1 = {f1_per_class[i]:.4f}")
        accuracies.append(acc)
        precisions.append(prec)
        recalls.append(rec)
        f1s.append(f1)
    mean_acc = np.mean(accuracies)
    std_acc = np.std(accuracies)
    print(f"\nMean Accuracy: {mean_acc:.4f}  {std_acc:.4f}")
    print(f"Mean Precision: {np.mean(precisions):.4f}")
    print(f"Mean Recall: {np.mean(recalls):.4f}")
    print(f"Mean F1-Score: {np.mean(f1s):.4f}")
    return mean_acc, precisions, recalls, f1s

# Main execution
if __name__ == "__main__":
    # Load SEED-VIG data
    try:
        X, y, features_per_step, time_steps = load_seed_vig_data("emotion_classified_seed_vig_dataset (2).csv")
        input_shape = (time_steps, features_per_step)
        print(f"Final input shape: {input_shape}")
    except Exception as e:
        print(f"Error loading data: {e}")
        raise ValueError("Failed to load CSV. Check file path, feature count, and column structure.")

    # Evaluate adapter-enhanced model
    print("Evaluating Adapter-Enhanced CNN-LSTM-Attention Model")
    mean_acc, precisions, recalls, f1s = evaluate_model(X, y, input_shape)

    # Compare with original study baselines
    baseline_models = {
        "SVM": 92.32,
        "Random Forest": 95.41,
        "XGBoost": 96.41,
        "MLP": 97.14,
        "Stacked Ensemble": 95.56
    }
    print("\nBaseline Accuracies from Original Study:")
    for model, acc in baseline_models.items():
        print(f"{model}: {acc:.2f}%")
    print(f"Adapter-Enhanced Model: {mean_acc*100:.2f}%")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import numpy as np
import pandas as pd
import keras
import seaborn as sns
import matplotlib.pyplot as plt

# Define constants
NUM_CLASSES = 7  # Seven emotion classes
ADAPTER_DIM = 32  # Increased bottleneck dimension for better representation
N_SPLITS = 5  # For 5-fold cross-validation
EMOTION_MAPPING = {
    "Angry": 0,
    "Calm": 1,
    "Frustrated": 2,
    "Happy": 3,
    "Neutral": 4,
    "Relaxed": 5,
    "Sad": 6
}

# Custom Attention Layer
class AttentionLayer(layers.Layer):
    def __init__(self, name="attention"):
        super(AttentionLayer, self).__init__(name=name)

    def call(self, inputs):
        lstm_output, attention_weights = inputs
        weighted_output = lstm_output * attention_weights
        context_vector = keras.ops.sum(weighted_output, axis=1)
        return context_vector

# Adapter Layer (Bottleneck MLP)
class AdapterLayer(layers.Layer):
    def __init__(self, input_dim, bottleneck_dim, name="adapter"):
        super(AdapterLayer, self).__init__(name=name)
        self.down_project = layers.Dense(bottleneck_dim, activation="relu")
        self.up_project = layers.Dense(input_dim)
        self.batch_norm = layers.BatchNormalization()

    def call(self, inputs, training=False):
        x = self.down_project(inputs)
        x = self.batch_norm(x, training=training)
        x = self.up_project(x)
        return x + inputs

# Custom Weighted Adapter Layer
class WeightedAdapterLayer(layers.Layer):
    def __init__(self, time_steps, name="weighted_adapter"):
        super(WeightedAdapterLayer, self).__init__(name=name)
        self.time_steps = time_steps

    def call(self, inputs):
        adapter_out, weight = inputs
        weight_expanded = keras.ops.expand_dims(weight, axis=1)
        weight_expanded = keras.ops.repeat(weight_expanded, self.time_steps, axis=1)
        return adapter_out * weight_expanded

# Gating Mechanism
class GatingMechanism(layers.Layer):
    def __init__(self, num_classes, num_adapters, name="gating"):
        super(GatingMechanism, self).__init__(name=name)
        self.num_classes = num_classes
        self.num_adapters = num_adapters
        self.emotion_predictor = layers.Dense(num_classes, activation="softmax", name="emotion_predictor")
        self.adapter_selector = layers.Dense(num_adapters, activation="softmax", name="adapter_selector")

    def call(self, inputs, training=False):
        emotion_dist = self.emotion_predictor(inputs)
        adapter_weights = self.adapter_selector(emotion_dist)
        return adapter_weights

# CNN-LSTM-Attention Model with Adapters
def build_cnn_lstm_attention_with_adapters(input_shape, num_classes=NUM_CLASSES):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv1D(filters=128, kernel_size=3, padding="same", activation="relu")(inputs)  # Increased filters
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    reduced_time_steps = input_shape[0] // 2
    cnn_adapters = [AdapterLayer(input_dim=128, bottleneck_dim=ADAPTER_DIM, name=f"cnn_adapter_{i}") for i in range(num_classes)]
    cnn_adapter_outputs = [adapter(x) for adapter in cnn_adapters]
    cnn_gating_input = layers.GlobalAveragePooling1D()(x)
    cnn_gating = GatingMechanism(num_classes=num_classes, num_adapters=num_classes, name="cnn_gating")
    cnn_adapter_weights = cnn_gating(cnn_gating_input)
    cnn_output = 0
    for i, adapter_out in enumerate(cnn_adapter_outputs):
        weighted_out = WeightedAdapterLayer(time_steps=reduced_time_steps, name=f"cnn_weighted_adapter_{i}")(
            [adapter_out, cnn_adapter_weights[:, i:i+1]]
        )
        cnn_output += weighted_out
    x = layers.LSTM(256, return_sequences=True, return_state=False)(cnn_output)  # Increased units, deeper stack
    x = layers.LSTM(128, return_sequences=True)(x)
    x = layers.BatchNormalization()(x)
    lstm_adapters = [AdapterLayer(input_dim=128, bottleneck_dim=ADAPTER_DIM, name=f"lstm_adapter_{i}") for i in range(num_classes)]
    lstm_adapter_outputs = [adapter(x) for adapter in lstm_adapters]
    lstm_gating_input = layers.GlobalAveragePooling1D()(x)
    lstm_gating = GatingMechanism(num_classes=num_classes, num_adapters=num_classes, name="lstm_gating")
    lstm_adapter_weights = lstm_gating(lstm_gating_input)
    lstm_output = 0
    for i, adapter_out in enumerate(lstm_adapter_outputs):
        weighted_out = WeightedAdapterLayer(time_steps=reduced_time_steps, name=f"lstm_weighted_adapter_{i}")(
            [adapter_out, lstm_adapter_weights[:, i:i+1]]
        )
        lstm_output += weighted_out
    attention_scores = layers.Dense(1, activation="tanh")(lstm_output)
    attention_weights = layers.Softmax(axis=1)(attention_scores)
    context_vector = AttentionLayer()([lstm_output, attention_weights])
    x = layers.Dense(256, activation="relu")(context_vector)  # Increased dense units
    x = layers.Dropout(0.3)(x)  # Adjusted dropout
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = Model(inputs, outputs)
    return model

# Function to find a suitable TIME_STEPS
def find_time_steps(num_features, preferred_divisors=[94, 47, 4, 85, 25, 5]):
    for divisor in preferred_divisors:
        if num_features % divisor == 0:
            return divisor
    for i in range(1, num_features + 1):
        if num_features % i == 0:
            return i
    raise ValueError(f"No valid TIME_STEPS found for {num_features} features.")

# Load and preprocess SEED-VIG dataset from CSV
def load_seed_vig_data(file_path="emotion_classified_seed_vig_dataset (2).csv"):
    try:
        data = pd.read_csv(file_path)
    except FileNotFoundError:
        raise FileNotFoundError(f"Dataset file {file_path} not found. Ensure the SEED-VIG CSV is in the working directory.")

    if "emotion" not in data.columns:
        raise ValueError("CSV must contain an 'emotion' column with string labels (Angry, Calm, Frustrated, Happy, Neutral, Relaxed, Sad).")

    if not all(data["emotion"].isin(EMOTION_MAPPING.keys())):
        invalid_labels = set(data["emotion"]) - set(EMOTION_MAPPING.keys())
        raise ValueError(f"Invalid emotion labels found: {invalid_labels}. Expected: {list(EMOTION_MAPPING.keys())}.")

    y = data["emotion"].map(EMOTION_MAPPING).values
    print(f"Emotion distribution (raw):\n{data['emotion'].value_counts()}")

    # Extract feature columns
    feature_columns = [col for col in data.columns if col not in ["vigilance", "valence", "arousal", "emotion", "binary_vigilance"]]
    num_features = len(feature_columns)
    print(f"Detected {num_features} feature columns: {feature_columns[:10]}...")
    print(f"Total columns in CSV: {len(data.columns)}")
    X = data[feature_columns].values
    print(f"Feature matrix shape: {X.shape}")

    # Check for expected EEG and ICA features
    expected_eeg = [f"{feat}_ch{ch}_band{band}" for feat in ["psd_movingAve", "psd_LDS", "de_movingAve", "de_LDS"]
                    for ch in range(17) for band in range(5)]
    expected_ica = [f"features_table_ica_{i}" for i in range(36)]
    expected_features = expected_eeg + expected_ica  # 340 + 36 = 425
    missing_features = [col for col in expected_features if col not in feature_columns]
    print(f"Missing {len(missing_features)} expected features: {missing_features[:10]}..." if missing_features else "All expected features present.")
    if num_features != 425:
        print(f"WARNING: Expected 425 features, but found {num_features}. Proceeding with {num_features} features.")

    # Dynamically select TIME_STEPS
    time_steps = find_time_steps(num_features)
    print(f"Selected TIME_STEPS={time_steps} for {num_features} features.")

    from sklearn.impute import SimpleImputer
    imputer = SimpleImputer(strategy="mean")
    X = imputer.fit_transform(X)
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    smote = SMOTE(random_state=42, sampling_strategy="auto")
    X_balanced, y_balanced = smote.fit_resample(X, y)
    features_per_step = num_features // time_steps
    X_reshaped = X_balanced.reshape(-1, time_steps, features_per_step)
    y_one_hot = tf.keras.utils.to_categorical(y_balanced, num_classes=NUM_CLASSES)
    print(f"Loaded data: {X_reshaped.shape} samples, {y_one_hot.shape} labels")
    print(f"Input shape: ({time_steps}, {features_per_step})")
    print(f"Balanced class distribution: {np.sum(y_one_hot, axis=0)}")
    return X_reshaped, y_one_hot, features_per_step, time_steps

# Evaluate model with cross-validation
def evaluate_model(X, y, input_shape, num_classes=NUM_CLASSES, n_splits=N_SPLITS):
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    accuracies = []
    precisions, recalls, f1s = [], [], []
    y_classes = y.argmax(axis=1)
    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y_classes)):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        model = build_cnn_lstm_attention_with_adapters(input_shape)
        model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4), loss="categorical_crossentropy", metrics=["accuracy"])
        # Callbacks for optimization
        early_stopping = EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)
        lr_scheduler = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6)
        model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.2,
                  callbacks=[early_stopping, lr_scheduler], verbose=1)
        y_pred = model.predict(X_test)
        y_pred_classes = y_pred.argmax(axis=1)
        y_test_classes = y_test.argmax(axis=1)
        acc = accuracy_score(y_test_classes, y_pred_classes)
        prec, rec, f1, _ = precision_recall_fscore_support(y_test_classes, y_pred_classes, average="weighted")
        prec_per_class, rec_per_class, f1_per_class, _ = precision_recall_fscore_support(
            y_test_classes, y_pred_classes, labels=range(num_classes)
        )
        print(f"Fold {fold+1}: Accuracy = {acc:.4f}, Precision = {prec:.4f}, Recall = {rec:.4f}, F1 = {f1:.4f}")
        print("Per-class F1-scores:")
        for i, emotion in enumerate(EMOTION_MAPPING.keys()):
            print(f"  {emotion}: F1 = {f1_per_class[i]:.4f}")
        # Confusion Matrix
        cm = confusion_matrix(y_test_classes, y_pred_classes)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=EMOTION_MAPPING.keys(), yticklabels=EMOTION_MAPPING.keys())
        plt.title(f"Confusion Matrix - Fold {fold+1}")
        plt.ylabel("True Label")
        plt.xlabel("Predicted Label")
        plt.show()
        accuracies.append(acc)
        precisions.append(prec)
        recalls.append(rec)
        f1s.append(f1)
    mean_acc = np.mean(accuracies)
    std_acc = np.std(accuracies)
    print(f"\nMean Accuracy: {mean_acc:.4f}  {std_acc:.4f}")
    print(f"Mean Precision: {np.mean(precisions):.4f}")
    print(f"Mean Recall: {np.mean(recalls):.4f}")
    print(f"Mean F1-Score: {np.mean(f1s):.4f}")
    return mean_acc, precisions, recalls, f1s

# Main execution
if __name__ == "__main__":
    # Load SEED-VIG data
    try:
        X, y, features_per_step, time_steps = load_seed_vig_data("emotion_classified_seed_vig_dataset (2).csv")
        input_shape = (time_steps, features_per_step)
        print(f"Final input shape: {input_shape}")
    except Exception as e:
        print(f"Error loading data: {e}")
        raise ValueError("Failed to load CSV. Check file path, feature count, and column structure.")

    # Evaluate adapter-enhanced model
    print("Evaluating Adapter-Enhanced CNN-LSTM-Attention Model")
    mean_acc, precisions, recalls, f1s = evaluate_model(X, y, input_shape)

    # Compare with original study baselines
    baseline_models = {
        "SVM": 92.32,
        "Random Forest": 95.41,
        "XGBoost": 96.41,
        "MLP": 97.14,
        "Stacked Ensemble": 95.56
    }
    print("\nBaseline Accuracies from Original Study:")
    for model, acc in baseline_models.items():
        print(f"{model}: {acc:.2f}%")
    print(f"Adapter-Enhanced Model: {mean_acc*100:.2f}%")
```
